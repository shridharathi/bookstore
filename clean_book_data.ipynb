```python
for i in range(len(curr_books)):
    curr_books['Category'].iloc[i] = str(curr_books['Category'].iloc[i])
    if curr_books['Category'].iloc[i] != 'NaN':
        curr_books['Category'].iloc[i] = curr_books['Category'].iloc[i].split(' ,')
    else:
        curr_books['Category'].iloc[i] = []
        
    curr_books['Description'].iloc[i] = str(curr_books['Description'].iloc[i])
    if curr_books['Description'].iloc[i] != 'NaN':
        curr_books['Description'].iloc[i] = ""
```

```python
for i in range(len(curr_books)):
    if curr_books['Category'].iloc[i][0] == "nan":
        curr_books['Category'].iloc[i] = []
```

```python
for i in range(len(curr_books)):
    if not curr_books['Description'].iloc[i]:
        curr_books['Description'].iloc[i] = curr_books['Title'].iloc[i]
```

```python
for i in range(len(curr_books)):
    if not curr_books['Category'].iloc[i] == []:
        curr_books['Category'].iloc[i] = curr_books['Category'].iloc[i][0]
    else: 
        curr_books['Category'].iloc[i] = ""
```

```python
def get_embedding(chunk):
  url = 'https://api.openai.com/v1/embeddings'
  headers = {
      'content-type': 'application/json; charset=utf-8',
      'Authorization': f"Bearer {OPENAI_API_KEY}"            
  }
  data = {
      'model': OPENAI_EMBEDDING_MODEL,
      'input': chunk
  }
  response = requests.post(url, headers=headers, data=json.dumps(data))  
  response_json = response.json()
  embedding = response_json["data"][0]["embedding"]
  return embedding
```
